```python
import wordsegment as ws

from wordsegment import load, segment

load()
  
def flat(l):
    for k in l:
        if not isinstance(k, (list, tuple)):
            yield k
        else:
            yield from flat(k)
```

### 1.3 Splitting

```python
# split every characters
data["simplesplit"] = ""
# split special characters
data["specialcharsplit"] = ""
# split words but don't remain the captial characters
data["extrawordsplit"] = ""



for index, row in data.iterrows():
    # simple split for each row
    row['simplesplit'] = list(row['original'])
    row['specialcharsplit'] = re.findall(r'[0-9]+|[a-zA-Z]+|[^a-zA-Z0-9]+',row['original'])
    
    passwdstring=[]
    for i in row['specialcharsplit']:
        templist = segment(i)
        if(templist):
            passwdstring.append(templist)
        else:
            passwdstring.append(i)
    row['extrawordsplit'] = list(flat(passwdstring))
```

```python
def insertChar(mystring, position, chartoinsert ):
    longi = len(mystring)
    mystring   =  mystring[:position] + chartoinsert + mystring[position:] 
    return mystring
```

```python
# split words and remain the captial charaters (fix captial issues)
data["fixcapital"] = ""

for index, row in data.iterrows():
    temp = row['extrawordsplit']
    mixwithspace = " ".join(temp)
    origin = row['original']
    for i in range(0,len(mixwithspace)):
        if(mixwithspace[i]==" "):
            origin = insertChar(origin,i," ")
    row["fixcapital"] = origin
```

### Try LDA with words only

```python
def LDA_words(docs_raw,n_topics):
    tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                    stop_words = 'english',
                                    lowercase = True,
                                    # token choose the words with more than 4 length and get rid of the speical words
                                    token_pattern = r'\b[a-zA-Z]{3,}\b',
                                    max_df = 0.5, 
                                    min_df = 10)
    dtm_tf = tf_vectorizer.fit_transform(docs_raw)
    tfidf_vectorizer = TfidfVectorizer(**tf_vectorizer.get_params())
    dtm_tfidf = tfidf_vectorizer.fit_transform(docs_raw)
    # for TF DTM
    lda_tf = LatentDirichletAllocation(n_components=n_topics, random_state=0)
    %time lda_tf.fit(dtm_tf)
    # for TFIDF DTM
    lda_tfidf = LatentDirichletAllocation(n_components=n_topics, random_state=0)
    %time lda_tfidf.fit(dtm_tfidf)
    return(lda_tf, dtm_tf, tf_vectorizer)

def LDA_words_numbers(docs_raw,n_topics):
    tf_vectorizer = CountVectorizer(strip_accents = 'unicode',
                                    stop_words = 'english',
                                    lowercase = True,
                                    # token choose the words with more than 4 length and get rid of the speical words
                                    token_pattern = r'\b[a-zA-Z0-9]{3,}\b',
                                    max_df = 0.5, 
                                    min_df = 10)
    dtm_tf = tf_vectorizer.fit_transform(docs_raw)
    tfidf_vectorizer = TfidfVectorizer(**tf_vectorizer.get_params())
    dtm_tfidf = tfidf_vectorizer.fit_transform(docs_raw)
    # for TF DTM
    lda_tf = LatentDirichletAllocation(n_components=n_topics, random_state=0)
    %time lda_tf.fit(dtm_tf)
    # for TFIDF DTM
    lda_tfidf = LatentDirichletAllocation(n_components=n_topics, random_state=0)
    %time lda_tfidf.fit(dtm_tfidf)
    return(lda_tf, dtm_tf, tf_vectorizer)

```
