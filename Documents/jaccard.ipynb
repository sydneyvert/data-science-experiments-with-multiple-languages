```python
import itertools 

def jaccard(a, b):
    a = set(list(a))
    b = set(list(b))
    c = a.intersection(b)
    return float(len(c)) / (len(a) + len(b) - len(c))
```

```python
str1 = 'hello'
str2 = 'no'

jaccard(str1, str2)
```
